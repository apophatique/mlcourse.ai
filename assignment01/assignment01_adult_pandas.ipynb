**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
# Ваш код здесь
columns_to_show = ['age']

data.groupby(['race', 'sex'])[columns_to_show].describe()
```

Ответ: 80

**8. Среди кого больше доля зарабатывающих больше (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# Ваш код здесь

married_num = len(data[data['marital-status'].apply(lambda x: 'Married' in x) & (data['salary'] == '>50K')])
not_married_num = len(data[data['marital-status'].apply(lambda x: 'Married' not in x) & (data['salary'] == '>50K')])

print(married_num > not_married_num)
```

Ответ: женатых мужчин с зарплатой больше 50К больше

**9. Какое максимальное количество часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
# Ваш код здесь
max_working_hours = data['hours-per-week'].describe()['max']
print("Максимальное количество часов работы в неделю", max_working_hours)
```

```python
max_working_hours_count = data[data['hours-per-week'] == max_working_hours]['age'].count()
print('Количество людей, работающих по 99 часов в неделю - ', max_working_hours_count)
```

```python
print('Среди них процент людей, зарабатывающих много, - ', \
      data[(data['hours-per-week'] == max_working_hours) & (data['salary'] == '>50K')]['age'].count()/max_working_hours_count
     )
```
